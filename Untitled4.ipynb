# Pipeline de Preparação dos Dados | census_Base [Fase 1]

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
label_encoder = LabelEncoder()
Y_census_treinamento_num = label_encoder.fit_transform(Y_census_treinamento)
```

```python
parametros = {'n_neighbors': [3, 5, 10, 20],
              'p': [1, 2]}
```

```python
grid_census = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=parametros)
```

```python
grid_census.fit(X_census_treinamento, Y_census_treinamento_num)
```

```python
melhor_parametros_census = grid_census.best_params_
print("Melhores parâmetros encontrados:", melhor_parametros_census)
```

```python
melhor_knn = grid_census.best_estimator_
previsoes_brutas = melhor_knn.predict(X_census_teste)
```

```python
previsoes_knn = label_encoder.inverse_transform(previsoes_brutas)
```

```python
acuracia_knn = accuracy_score(Y_census_teste, previsoes_knn)
print(f"Acurácia do KNN na base de teste: {acuracia_knn:.4f}")
```

```python
print("\nRelatório de Classificação do KNN:")
print(classification_report(Y_census_teste, previsoes_knn))
```
